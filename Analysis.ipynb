```python
lst_col = 'genre'
```

```python
x = Movies_df.assign(**{lst_col:Movies_df[lst_col].str.split(',')})
```

```python

#https://stackoverflow.com/questions/12680754/split-explode-pandas-dataframe-string-entry-to-separate-rows
Movies_New = pd.DataFrame({                         
col:np.repeat(x[col].values, x[lst_col].str.len())
for col in x.columns.difference([lst_col])
}).assign(**{lst_col:np.concatenate(x[lst_col].values)})[x.columns.tolist()]
 
```

```python
Movies_New.head()
```

```python
Movie_Rating = Movies_New[['imdb_title_id','title','date_published','genre','votes']]
Movie_Rating
```

```python
#Split the date column in the df in order to get the years.
Movie_Rating[["Year_Published","Months"]] = Movie_Rating['date_published'].str.split('-',n=1,expand=True)
```

```python
Movie_Rating.drop(['Months','date_published'],inplace=True,axis=1)
Movie_Rating.to_csv("Main Data/Movie_Demo.csv",index=False)
```

```python
Movie_Rating['genre'].value_counts()
```

```python
#Removed the extra spaces from the cells.
Movie_Rating['genre'] = Movie_Rating['genre'].str.replace(" ","")
Movie_Rating['genre'].value_counts()
```
